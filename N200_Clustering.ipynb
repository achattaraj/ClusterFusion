Timecourse of Collective variables (.colvars.traj files) are averaged across 5 runs and saved it as "Mean_CVtrajectory.txt". Average of potential mean force (.pmf files) is saved in "Mean_PMF.txt".

```python
import numpy as np
import matplotlib.pyplot as plt


file1 = "./N200_Clustering/MultiRun_data/Mean_CVtrajectory.txt"

d = np.loadtxt(file1)

plt.plot(d[:,0], d[:,1], lw=2)
plt.fill_between(d[:,0], d[:,1]-d[:,2], d[:,1]+d[:,2], alpha=0.3)
plt.ylim(80,180)
plt.ylabel('R$_g ^{system}$ ($\AA$)')
plt.xlabel('Time ($\mu$s)')
plt.show()
```

```python
file2 = "./N200_Clustering/MultiRun_data/Mean_PMF.txt"

d = np.loadtxt(file2)

plt.plot(d[:,0], d[:,1], lw=2)
plt.fill_between(d[:,0], d[:,1]-d[:,2], d[:,1]+d[:,2], alpha=0.3)

plt.xlabel('R$_g ^{system}$ ($\AA$)')
plt.ylabel('Potential Mean Force (kcal/mol)', fontweight='bold', fontsize=14)
plt.show()
```
